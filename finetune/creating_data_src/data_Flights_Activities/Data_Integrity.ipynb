```python
import json
from collections import Counter

def count_destinations(filename):
    with open(filename, 'r') as file:
        data = json.load(file)
        # Initialize a list to store the destinations or origins
        destinations = []
        # Assuming the JSON data is a list of such dictionaries
        for item in data:
            # Ensure the nested keys exist before accessing them
            if 'output' in item and 'action_input' in item['output'] and 'destination' in item['output']['action_input']:
                destinations.append(item['output']['action_input']['destination'])
        # Count each unique destination
        destination_counts = Counter(destinations)
        return destination_counts

# Example usage:
destination_counts = count_destinations('Modified_Modified_Modified_Modified_flight_activities.json')
print(destination_counts)

```

```python
import json
from datetime import datetime

def is_valid_date(date_str):
    """Check if the date string is in 'YYYY-MM-DD' format."""
    if not isinstance(date_str, str):
        return False
    try:
        datetime.strptime(date_str, '%Y-%m-%d')
        return True
    except ValueError:
        return False

def ensure_correct_type(value, expected_data_type):
    """
    Ensure that the value is of the expected data type.
    If it's not, attempt to convert or fix it.
    """
    if expected_data_type == int:
        if isinstance(value, int):
            return value
        elif isinstance(value, float):
            return int(value)
        elif isinstance(value, list) and len(value) == 1:
            return ensure_correct_type(value[0], int)
        elif isinstance(value, str) and value.isdigit():
            return int(value)
        else:
            return None  # Return None for invalid int to handle maxPrice removal
    elif expected_data_type == str:
        if isinstance(value, str):
            return value
        elif isinstance(value, (int, float)):
            return str(value)
        elif isinstance(value, list) and len(value) == 1:
            return ensure_correct_type(value[0], str)
        else:
            return ''
    elif expected_data_type == bool:
        if isinstance(value, bool):
            return value
        elif isinstance(value, str):
            if value.lower() == 'true':
                return True
            elif value.lower() == 'false':
                return False
            else:
                return False
        else:
            return False
    elif expected_data_type == list:
        if isinstance(value, list):
            return value
        else:
            return []
    else:
        return value

def process_data(data):
    """
    Process each entry in the data list:
    - Check for required fields and validate them.
    - If optional fields are present, ensure they are in the correct format.
    - Do not add missing optional fields.
    - Exclude entries with missing or invalid required fields.
    - Handle 'maxPrice' according to the new requirement.
    """
    required_fields = {
        'action': list,  # Under 'output'
        'origin': str,   # Under 'action_input'
        'destination': str,  # Under 'action_input'
        'departure_date': str,  # Under 'action_input'
        'adults': int,  # Under 'action_input'
        'max': int,  # Under 'action_input'
    }

    optional_fields = {
        'return_date': '',      # Expected type: str
        'children': 0,          # Expected type: int
        'infants': 0,           # Expected type: int
        'class': '',            # Expected type: str
        'nonStop': False,       # Expected type: bool
        # 'maxPrice' is intentionally omitted from optional_fields
        #'ratings': 0,           # Expected type: int
        #'amenities': [],        # Expected type: list
        # 'max' is already included in required_fields
    }

    output_data = []
    for obj in data:
        valid = True

        # Check if 'output' and 'action_input' are present
        if 'output' not in obj or 'action_input' not in obj['output']:
            print('Missing output or action_input:', json.dumps(obj, indent=4))
            continue  # Skip this object

        # Check 'action' in obj['output']
        if 'action' not in obj['output']:
            print(f"Missing required field 'action':", json.dumps(obj, indent=4))
            continue  # Skip this object
        else:
            action_value = obj['output']['action']
            # Ensure 'action' is exactly ["search_flights", "search_hotels", "activities"]
            if action_value != ["search_flights", "activities"]:
                print(f"Invalid 'action' field:", json.dumps(obj, indent=4))
                continue  # Skip this object

        action_input = obj['output']['action_input']

        # Ensure 'action_input' is a dictionary
        if not isinstance(action_input, dict):
            print('action_input is not a dictionary:', json.dumps(obj, indent=4))
            continue

        # Check required fields under 'action_input'
        for field, expected_data_type in required_fields.items():
            if field == 'action':
                continue  # Already checked
            if field not in action_input:
                print(f"Missing required field '{field}':", json.dumps(obj, indent=4))
                valid = False
                break
            else:
                value = action_input[field]
                # Ensure the value is of the correct type and format
                if field == 'departure_date':
                    if not is_valid_date(value):
                        print(f"Invalid 'departure_date' format:", json.dumps(obj, indent=4))
                        valid = False
                        break
                elif field in ['origin', 'destination']:
                    action_input[field] = ensure_correct_type(value, str)
                    if not action_input[field]:
                        print(f"Invalid '{field}' value:", json.dumps(obj, indent=4))
                        valid = False
                        break
                else:
                    # Ensure correct data type
                    action_input[field] = ensure_correct_type(value, expected_data_type)

        if not valid:
            continue  # Skip this object

        # For optional fields that are present, ensure they are in the correct format
        for field in optional_fields.keys():
            if field in action_input and action_input[field] is not None:
                expected_data_type = type(optional_fields[field])
                action_input[field] = ensure_correct_type(action_input[field], expected_data_type)
                # Additional date validation for 'return_date'
                if field == 'return_date' and action_input[field]:
                    if not is_valid_date(action_input[field]):
                        print(f"Invalid 'return_date' format in field '{field}':", json.dumps(obj, indent=4))
                        action_input[field] = ''  # Reset to empty string or None, as per your preference

        # Handle 'maxPrice' according to your requirements
        if 'maxPrice' in action_input:
            max_price_value = ensure_correct_type(action_input['maxPrice'], int)
            if max_price_value is None or max_price_value == 0:
                # Remove 'maxPrice' if it's 0 or invalid
                del action_input['maxPrice']
            else:
                # Keep 'maxPrice' if it's greater than 0
                action_input['maxPrice'] = max_price_value

        # Append the validated and fixed object to the output list
        output_data.append(obj)

    return output_data

def main():
    """Main function to process the JSON data."""
    # Read data from 'Flight_Hotel_Activ.json'
    with open('Flight_Activ.json', 'r') as f:
        data = json.load(f)

    # Process the data
    processed_data = process_data(data)

    # Write the processed data to 'new.json'
    with open('new_2.json', 'w') as f:
        json.dump(processed_data, f, indent=4)

    # Print the length of the new JSON
    print(f"Number of entries in new.json: {len(processed_data)}")

if __name__ == '__main__':
    main()
```
